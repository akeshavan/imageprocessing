## a. Create a function called loadview to load and display an image

```python
def rgb2gray(rgb):
    #This function is based on matlab's rgb2gray function
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])

def loadfig(img_path):
    img = rgb2gray(imread(img_path))
    return img
    
def loadview(img,ax=None):
    if not ax: fig, ax = subplots(1,1)
    ax.imshow(img,cmap=gray())
    ax.set_xticklabels('')
    ax.set_yticklabels('')
    return ax
```

## b. Perform each of these procedures on the original image.

### Subsample the image by 2 in each dimension.

```python
def subsample2(img):
    shape = img.shape
    new_shape = [s/2 for s in shape]
    new_img = zeros(new_shape)
    for i,idx in enumerate(range(0,shape[0],2)):
        for j,jdx in enumerate(range(0,shape[1],2)):
            new_img[i,j] = img[idx,jdx]
    return new_img
```

```python
new_img = subsample2(img)
loadview(new_img);
print img.shape, "subsampled to", new_img.shape
```

### Crop the image to the center 60 pixels

```python
def centercrop(img,num=60):
    shape = img.shape
    center = [s/2 for s in shape]
    xmin, xmax = center[0]-num/2, center[0]+num/2
    ymin, ymax = center[1]-num/2, center[1]+num/2
    return img[xmin:xmax,ymin:ymax]
```

```python
img_crop = centercrop(img)
loadview(img_crop);
```

### Flip the image vertically so that the bottom row becomes the top row and the top row  becomes the bottom row

```python
def flip_vert(img):
    return img[::-1,:]
```

```python
img_flipvert = flip_vert(img)
loadview(img_flipvert);
```

### Flip the image horizontally so that the right column becomes the left column and vice versa 

```python
def flip_horiz(img):
    return img[:,::-1]
```

```python
img_fliph = flip_horiz(img)
loadview(img_fliph);
```

### Invert the intensities in the image so that black pixels become white and vice versa. DO  NOT CREATE A BINARY (ie black and white) IMAGE. It should be a grayscale image  that looks like the negative of the original images.

```python
def invert(img):
    imax = np.max(img)
    img2 = np.asarray([imax-i for i in img.ravel()])
    img2 = img2.reshape(img.shape)
    return img2
```

```python
img_invert = invert(img)
loadview(img_invert);
```

```python
def mat2grey(mat,scale = None):
    imin,imax = np.min(mat), np.max(mat)
    # set all values higher than amax to one, all vals lower than a min to 0, and everything inbetween to range(0,1)
    
    if not scale:
        amin, amax = imin, imax
    else:
        amin = np.min(scale)
        amax = np.max(scale)
    newmat = zeros(mat.shape)
    newmat[mat >= amax] = 1
    newmat[mat <= amin] = 0
    idx = (np.multiply((mat>= amin),(mat<= amax))).astype(bool)
    newmat[idx] = (mat[idx]-amin)/float(amax-amin)
    return newmat
    
```

```python
mat2grey(f,[1,255])
```

```python
mat2grey(g,[1,255])
```

####Comments####

* the values of doubles (f) get converted to values in the range (0,1) just fine

* since g is uint 8, the math is messed up because integer division isn't true division like we are used to

## Quantize

```python
def quantize(x, b):
    #Y = QUANTIZE(X,B) quantizes X to B bits. The quantized bits are in the
    #upper half of the byte (ie 128, 64, etc) to aid in the display of the
    #quantized image.
    #Create bit masks for the quantization

    lo = uint8(2**(8 - b) - 1)
    hi = uint8(2**8 - double(lo) - 1)
    
    #Perform standard quantization
    y = np.bitwise_and(x.astype(int), hi)
    return y
```

####Comments####

* bitwise and converts the first and second inputs to binary, 
* For each digit of each binary number, if the number match then it adds 2^k where k is the digit #
* it returns the final sum after comparing each digit in the binary number
* For example, 2 in binary is 10 and 3 is 11, so bitwise and should return 1*2^2 + 0*2^1 = **2** since they match at the first digit (1 and 1) and don't match at the last digit (0 and 1)
* The bitwise and only works if the input matrix consists of integers
* in order to reduce the number of colors to 2^1 


```python
np.bitwise_and(2,3)
```
